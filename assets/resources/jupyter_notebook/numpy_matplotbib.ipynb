Multiply two long arrays

For this, let's take to very long arrays (e.g., 1 millon or 10 millon elements) and multiply them element-wise. Mathematically, we want  ci=ai‚àóbi . We will do this with a for loop and with NumPy's vectorized implementation and compare their wallclock times using  ùöùùöíùöñùöé.ùöùùöíùöñùöé() .

```python
a = np.random.randn(10000000)
b = np.random.randn(10000000)
```

```python
c = np.empty_like(a)
start = time.time()
for i in range(len(a)):
    c[i] = a[i] * b[i]
end = time.time()
print(end - start)
```

```python
start = time.time()
c = a * b
end = time.time()
print(end - start)
```

```python
start = time.time()
s = sum(a * b)
end = time.time()
print(end - start)
```

```python
start = time.time()
np.expand_dims(a, 1).T.dot(b)
end = time.time()
print(end - start)
```

```python
start = time.time()
np.inner(a, b)
end = time.time()
print(end - start)
```

FFT PLotting

Let's now use NumPy and Matplotlib together to visualize the Fourier transform of a sine function.

```python
t = np.arange(256)
a = np.sin(t)
A = np.fft.fft(a)
f = np.fft.fftfreq(t.shape[-1])
```

```python
plt.subplot(2, 1, 1)
plt.title('Time series')
plt.plot(t, a)
plt.ylabel('sin(t)')
plt.xlabel('t')
plt.subplot(2,1,2)
plt.plot(f, A.real, f, A.imag)
plt.ylabel('Power')
plt.xlabel('f')
plt.legend(['Real', 'Imag'])
plt.tight_layout()
plt.show()
```
